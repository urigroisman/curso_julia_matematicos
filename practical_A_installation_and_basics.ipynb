### Start with using the REPL

1. Run the Julia REPL.

```
for i in 1:5
    println("Hello $i")
end
```
1. Now enter it again, changing `$i` to `$(i % 3)`. Explain why you get this output. Note that you can use the up arrow to change the command prompt back to the previous command so you don't have to enter the whole thing again.

6. Now hit `;` to enter shell mode.
