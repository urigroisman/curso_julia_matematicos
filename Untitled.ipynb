Supongamos que tenemos un arreglo (array) de números y queremos hallar el valor máximo.  

Podemos hacerlo usando la función predefinida (in-built function) `maximum`
Siempre es bueno porque están optimizadas

```julia
using Random
Random.seed!(0)

data = rand(10^8)

maximum(data)
```

o programando una nuestra (siempre más lenta)

```julia
function find_max(data)
    max_value = -Inf
    for x in data       #esta forma de for loop, recorre todos los elementops del array (data)
        if x > max_value
            max_value = x
        end
    end
    return max_value    #= si no aclaramos que devolvemos, la función siempre devuelve
                        el último valor calculado, que npo siempre es fácil saber cual es.  
                        Además es factible devolver más de un valor =#
end
```

```julia
Random.seed!(7)
data = rand(10^8)
@time max_value_in_built = maximum(data)
@time max_value = find_max(data)
println("Found maximal value: $(max_value)","  ", "$(max_value_in_built)")
```

<br><br>
Ahora queremos saber **cuantas veces**, mientras recorremos el array encontramos un máximo relativo.  
  Lo hacemos agregandooslos un para de lineas a nuestra función y devolvemos 2 valores

```julia
function find_max(data)
    max_value = -Inf
    count = 0
    for x in data
        if x > max_value
            count = count + 1
            max_value = x
        end
    end
    return max_value, count
end
```

```julia
Random.seed!(3)
data = rand(10^8)
max_value, n = find_max(data)
println("Found maximal value: $(max_value)")
println("There were $n records along the way.")
```

<br><br>
Nos podríamos preguntar cuantos máximos relativos hay en promedio cuando n se hace más y más grande.  Este es un problema no trivial en propbabilidad, pero existe una fórmula desarrollada por [Euler-Mascheroni](https://en.wikipedia.org/wiki/Euler–Mascheroni_constant) que depende de una constante **$\gamma$** que al igual que muchas constantes matemáticas y físicas están incluidas en `Julia`


$$
X_n = I_1 + I_2 + \dots + I_n,
$$

donde,

$$
I_i =
\begin{cases} 
1 & \text{if the } i\text{'th data point is a record}, \\
0 & \text{otherwise}.
\end{cases}
$$

Ahora, el valor esperado $\mathbb{E}[X_n]$ es,

$$
\mathbb{E}[X_n] = h_n = \sum_{i=1}^{n} \frac{1}{i},
$$

es la suma armónica y sabemos que:

$$
h_n = \log(n) + \gamma + o(1),
$$

donde $(\gamma)$ es la constante de **Euler-Mascheroni** y $( o(1) $) es un término que se anula a 1er órden cuando $( n \to \infty $). Eso es,

$$
h_n \approx \log(n) + \gamma.
$$



```julia
using Plots

γ = MathConstants.γ   # la función devuelve el valor de γ como irracional

γ = Float64(γ)

println("γ = $γ")

#= 
Definición de h(n) en una función in-line que es equivalente a:
function h(n)
    for i in 1:n
       h_n = h_n + 1/i
    end
    return h_n
end

También defino otra función in-line para la aproximación lineal
=#

h(n) = sum(1/i for i in 1:n)

approx_h(n) = log(n) + MathConstants.γ

#=
A continuación dos formas diferentes de llamar un for loop y en una de ellas
imprimimos datos de h(n) y de la aproximación lineal y en la segunda 
construimos el array err con la diferencia entre los dos valores para n de 1 a 20
=#

for n in 1:10
    println(n, "\t", "\t", h(n), "\t","\t", approx_h(n) )
end

err = [h(n) - approx_h(n) for n in 1:20]

# sintaxis para graficar una gráfica tipo scatter
scatter(1:20,err,xticks=1:20,xlabel="n",ylabel = "Error",ylim = (0,0.5), legend=false)
```

estimación de Monte Carlo y un histograma de la distribución, mismo tema pero sin usar la aproximación para h(n):

```julia
using Statistics

records = [] #Array vacío (records) y sin definir ni el tipo ni la dimensión

for s in 1:10^5      # 100000 muestras para el método Monte Carlo
    Random.seed!(s)  # de esta forma cada pasada se generan 100 números aleatorios diferentes
    data = rand(100)
    _, n = find_max(data) #uso mi función find_max que devuelve 2 valores
    push!(records,n)      #incluimos esos valores en records 
end
println("log(100) + γ =  $(approx_h(100)), Monte Carlo Estimate: $(mean(records))")

histogram(records,ticks = 1:maximum(records),normed=true,xlabel="n",ylabel="probability",legend=false)
```

## Bits, bytes, hex, binary, and the works...

A **bit** is either 0 or 1. A **byte** is a sequence of 8 bits. Hence there are $2^8=256$ options.

If you use a byte to encode an integer, then one option is the range $0,1,2,…,255$. This is called an **unsigned** integer, or `UInt8` in `Julia`.

Another option is the range $−127,…,128$ (still 256 possibilities) and this is called an `Int8`.

```julia
println("Unsigned 8 bit integers\n") 

println("Number \t Binary \tHex")

for n in 0:255
    
    # imprimo los primeros 18 y los últimos 6 números y 3 puntitos para que quede lindo
    # n en base 10, en representación binaria (función bitstring) y en HEX, base 16
    if n ≤ 18 || n ≥ 250 #\le + [TAB],  \ge + [TAB]
        println(n, ":\t", bitstring(UInt8(n)), "\t", string(n, base = 16))
    elseif n ∈ 19:21 #in(n,19:21)  \in + [TAB]
        println("\t   .")
    end
    
end
```

```julia
println("Signed 8 bit integers (observe the sign bit)\n")
println("Number \t Binary")
for n in -128:127
    if n ∈ -128:-120 || n ∈ -5:5 || n ∈ 125:127
        println(n, ":\t", bitstring(Int8(n)))
    elseif n ∈ union(-119:-117,6:8)
        println("\t   .")
    end
end
```

```julia
half_adder(a::T, b::T) where {T <: Bool} = (xor(a,b), a & b)
 
function full_adder(a::T, b::T, carry::T) where {T<:Bool}
    (s, carry_a) = half_adder(carry, a)
    (s, carry_b) = half_adder(s, b)
    return (s, carry_a | carry_b)
end

function add_n_digits(a::T, b::T, n) where {T <: Int}
    #calculate y = a + b using array indexing and logic operations
    #convert to a vector of booleans
    a_bool = digits(Bool, a, base = 2, pad = n)
    b_bool = digits(Bool, b, base = 2, pad = n)
    y_bool = fill(Bool(0), n)
    carry = false
    for i = 1 : n
        (y_bool[i], carry) = full_adder(a_bool[i], b_bool[i], carry)
    end
    return y_bool, a_bool, b_bool
end

#turn boolean array back into a string
display_bool_array(y) = join(reverse(Int.(y)), "")

a = 1
b = 1

y_bool, a_bool, b_bool = add_n_digits(a, b, 16)

println("adding $a + $b")
println("a = $(display_bool_array(a_bool))")
println("b = $(display_bool_array(b_bool))")
println("y = $(display_bool_array(y_bool))", "\n")

a = 1
b = 2^16 - 1 #biggest 16 bit UInt

y_bool, a_bool, b_bool = add_n_digits(a, b, 16)

println("adding $a + $b")
println("a = $(display_bool_array(a_bool))")
println("b = $(display_bool_array(b_bool))")
println("y = $(display_bool_array(y_bool))", "\n")
```
